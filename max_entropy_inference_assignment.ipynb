# Statistical Mechanics of the Quantum Harmonic Oscillator

Consider a single quantum harmonic oscillator of angular frequency $\omega$ in thermal equilibrium with a heat bath at temperature $T$. The energy eigenvalues are

$$E_j = \hbar\omega\left(j+\frac{1}{2}\right), \qquad j = 0,1,2,\dots$$

Define the dimensionless variable $x \equiv \dfrac{\hbar\omega}{k_BT} = \beta\hbar\omega$, where $\beta = 1/k_BT$.

---

## 1. The Partition Function

$$Z(T) = \sum_{j=0}^{\infty} e^{-\beta E_j} = \sum_{j=0}^{\infty} e^{-\beta\hbar\omega(j+1/2)} = e^{-\beta\hbar\omega/2}\sum_{j=0}^{\infty}\left(e^{-\beta\hbar\omega}\right)^{j}$$

This is a geometric series with ratio $r = e^{-\beta\hbar\omega}$. Since $T>0$, $\omega>0$, we have $0<r<1$, so the series converges:

$$\sum_{j=0}^{\infty} r^j = \frac{1}{1-r}$$

Therefore:

$$Z(T) = \frac{e^{-\hbar\omega/2k_BT}}{1-e^{-\hbar\omega/k_BT}}$$

---

## 2. Thermodynamic Potentials

**Helmholtz Free Energy.** Taking the log of $Z$:

$$\ln Z = -\frac{\beta\hbar\omega}{2} - \ln\left(1-e^{-\beta\hbar\omega}\right)$$

$$F(T) = -k_BT\ln Z = \frac{\hbar\omega}{2} + k_BT\ln\left(1-e^{-\hbar\omega/k_BT}\right)$$

**Internal Energy.** Using $U = -\dfrac{\partial \ln Z}{\partial \beta}$:

$$U(T) = \frac{\hbar\omega}{2} + \frac{\hbar\omega}{e^{\hbar\omega/k_BT}-1}$$

The first term is the zero-point energy; the second is $\hbar\omega\langle n\rangle$ with $\langle n\rangle = (e^{\beta\hbar\omega}-1)^{-1}$ the Bose–Einstein occupation number.

---

## 3. Entropy

Differentiating $F(T)$ with respect to $T$ ($S = -\partial F/\partial T$):

$$S(T) = k_B\left[\frac{x}{e^{x}-1} - \ln\left(1-e^{-x}\right)\right]$$

**Check $U = F + TS$:**

$$F+TS = \frac{\hbar\omega}{2} + k_BT\ln(1-e^{-x}) + T\left\{k_B\frac{x}{e^x-1} - k_B\ln(1-e^{-x})\right\}$$

The log terms cancel, leaving $F+TS = \dfrac{\hbar\omega}{2} + \dfrac{\hbar\omega}{e^x-1} = U$ ✓

---

## 4. Heat Capacity and Hyperbolic Form

$$C_V = \frac{\partial U}{\partial T} = k_B\,\frac{x^2 e^{x}}{(e^{x}-1)^2}$$

Using $e^x - 1 = e^{x/2}(e^{x/2}-e^{-x/2})$ and $\sinh(x/2) = \dfrac{e^{x/2}-e^{-x/2}}{2}$:

$$C_V = k_B\left(\frac{\hbar\omega}{2k_BT}\right)^2\frac{1}{\sinh^2\!\left(\dfrac{\hbar\omega}{2k_BT}\right)}$$

---

## 5. Asymptotic Behavior: High-Temperature Limit

Let $u = \dfrac{\hbar\omega}{2k_BT}\to 0$ as $T\to\infty$. Since $\sinh(u)\approx u + u^3/6$:

$$C_V \approx k_B\left(1 - \frac{u^2}{3}\right) \longrightarrow k_B \quad \text{as } T\to\infty$$

This matches the **classical equipartition theorem**: two quadratic degrees of freedom (kinetic + potential) each contribute $\tfrac12 k_B$, giving $C_V = k_B$.

As $T\to 0$, $C_V \to 0$ exponentially (quantum freeze-out) — this was central to Einstein's resolution of the specific-heat puzzle of solids.

# Q. The Maximum Entropy Distribution under Mean and Variance Constraints

Consider the random variable $X \in (-\infty, \infty)$. Let:

1. **The Mean:** $\mathbb{E}[X] = \mu$
2. **The Variance:** $\mathbb{E}[(X-\mu)^2] = \sigma^2$

Show that the law of $X$ that maximizes entropy is given by:

$$X \sim \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

# The Maximum Entropy Distribution under Mean and Variance Constraints

**Problem.** Consider a random variable $X \in (-\infty,\infty)$ with probability density $f(x)$, subject to:

$$\int_{-\infty}^{\infty} f(x)\,dx = 1 \qquad \text{(normalization)}$$
$$\mathbb{E}[X] = \int_{-\infty}^{\infty} x\,f(x)\,dx = \mu \qquad \text{(mean)}$$
$$\mathbb{E}[(X-\mu)^2] = \int_{-\infty}^{\infty} (x-\mu)^2 f(x)\,dx = \sigma^2 \qquad \text{(variance)}$$

We want to find the $f(x)$ that **maximizes** the differential entropy

$$H[f] = -\int_{-\infty}^{\infty} f(x)\ln f(x)\,dx$$

subject to these three constraints.

---

## 1. Setting Up the Variational Problem

This is a constrained optimization over functions, solved using **calculus of variations with Lagrange multipliers**. We introduce multipliers $\lambda_0, \lambda_1, \lambda_2$ for the three constraints and form the functional:

$$\mathcal{L}[f] = -\int f(x)\ln f(x)\,dx - \lambda_0\left(\int f(x)\,dx - 1\right) - \lambda_1\left(\int x f(x)\,dx - \mu\right) - \lambda_2\left(\int (x-\mu)^2 f(x)\,dx - \sigma^2\right)$$

We can write this as a single integral:

$$\mathcal{L}[f] = \int \left[-f(x)\ln f(x) - \lambda_0 f(x) - \lambda_1 x f(x) - \lambda_2 (x-\mu)^2 f(x)\right] dx + \text{constants}$$

---

## 2. Taking the Functional Derivative

Define the integrand as $\mathcal{F}(x, f) = -f\ln f - \lambda_0 f - \lambda_1 x f - \lambda_2 (x-\mu)^2 f$.

Since $\mathcal{F}$ depends on $f$ but not on $f'$, the Euler–Lagrange condition reduces to simply setting the partial derivative with respect to $f$ to zero:

$$\frac{\partial \mathcal{F}}{\partial f} = -\ln f - 1 - \lambda_0 - \lambda_1 x - \lambda_2 (x-\mu)^2 = 0$$

Solving for $\ln f$:

$$\ln f(x) = -1 - \lambda_0 - \lambda_1 x - \lambda_2 (x-\mu)^2$$

Therefore:

$$f(x) = \exp\left[-(1+\lambda_0) - \lambda_1 x - \lambda_2(x-\mu)^2\right]$$

This has the general form of a **Gaussian** (an exponential of a quadratic in $x$). We now fix the multipliers using the three constraints.

---

## 3. Determining the Lagrange Multipliers

**Step 1: Complete the square.** Write the exponent as a quadratic in $x$, centered conveniently around $\mu$. Absorb constants into a normalization factor $A = e^{-(1+\lambda_0)}$:

$$f(x) = A\,\exp\left[-\lambda_1 x - \lambda_2 (x-\mu)^2\right]$$

**Step 2: Use symmetry.** For $f(x)$ to have mean exactly $\mu$, the distribution must be symmetric about $x=\mu$. This forces the linear term to vanish when we complete the square around $\mu$ — i.e., $\lambda_1$ must combine with the $-\lambda_2(x-\mu)^2$ term such that the maximum of $f$ sits at $x=\mu$. Formally, writing $x = (x-\mu) + \mu$:

$$-\lambda_1 x - \lambda_2(x-\mu)^2 = -\lambda_2(x-\mu)^2 - \lambda_1(x-\mu) - \lambda_1\mu$$

For the mean of this distribution to equal $\mu$ exactly, the linear term in $(x-\mu)$ must vanish (odd term integrates to zero only if there's no linear piece), so **$\lambda_1 = 0$**. This gives:

$$f(x) = A\,\exp\left[-\lambda_2(x-\mu)^2\right]$$

**Step 3: Normalization.** Using the Gaussian integral $\displaystyle\int_{-\infty}^{\infty} e^{-a t^2}\,dt = \sqrt{\pi/a}$:

$$\int_{-\infty}^{\infty} A\,e^{-\lambda_2(x-\mu)^2}\,dx = A\sqrt{\frac{\pi}{\lambda_2}} = 1 \quad\Rightarrow\quad A = \sqrt{\frac{\lambda_2}{\pi}}$$

**Step 4: Match the variance.** Using $\displaystyle\int_{-\infty}^{\infty} t^2 e^{-at^2}\,dt = \frac{1}{2a}\sqrt{\frac{\pi}{a}}$:

$$\mathbb{E}[(X-\mu)^2] = A\int_{-\infty}^{\infty}(x-\mu)^2 e^{-\lambda_2(x-\mu)^2}\,dx = A\cdot\frac{1}{2\lambda_2}\sqrt{\frac{\pi}{\lambda_2}} = \frac{1}{2\lambda_2}$$

Setting this equal to $\sigma^2$:

$$\frac{1}{2\lambda_2} = \sigma^2 \quad\Rightarrow\quad \lambda_2 = \frac{1}{2\sigma^2}$$

Substituting back into $A$:

$$A = \sqrt{\frac{\lambda_2}{\pi}} = \sqrt{\frac{1}{2\pi\sigma^2}} = \frac{1}{\sqrt{2\pi\sigma^2}}$$

---

## 4. Final Result

Putting it all together:

$$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}}\exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)$$

which is exactly the **Gaussian (Normal) distribution**:

$$\boxed{X \sim \mathcal{N}(\mu,\sigma^2), \qquad f(x) = \frac{1}{\sqrt{2\pi\sigma^2}}\exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)}$$

---

## 5. Verifying This Is a Maximum (Not a Minimum or Saddle)

The second variation of the entropy functional is:

$$\delta^2 H = -\int \frac{(\delta f)^2}{f}\,dx \le 0$$

for any admissible perturbation $\delta f$, since $f(x) > 0$ everywhere. This confirms the stationary point found above is indeed a **maximum** of the entropy, not a minimum or saddle point — entropy is a strictly concave functional of $f$ (given fixed moments), so the maximizer is unique.

---

## 6. Interpretation

This result is the continuous analogue of a general principle: **among all distributions with a given mean and variance, the Gaussian has the maximum entropy** — i.e., it is the "least informative" or "maximally uncertain" distribution consistent with only knowing the first two moments. This is part of why the Gaussian appears so ubiquitously in physics and information theory: if you only know a system's mean and variance (e.g., from macroscopic thermodynamic measurements) and want to make the fewest additional assumptions, the Gaussian is the natural choice (this is the guiding principle behind Jaynes' **Maximum Entropy Principle** in statistical mechanics).

# Maximum Entropy Distribution for Asset Pricing

## Problem Setup

We seek the continuous density $p(S)$, defined for $S>0$ (asset price cannot be negative), that **maximizes entropy**

$$H[p] = -\int_0^{\infty} p(S)\ln p(S)\,dS$$

subject to four constraints:

$$\text{(Normalization)} \qquad \int_0^{\infty} p(S)\,dS = 1$$
$$\text{(Forward price)} \qquad \mathbb{E}[S_T] = \int_0^{\infty} S\,p(S)\,dS = 102$$
$$\text{(Volatility)} \qquad \mathbb{E}[S_T^2] = \int_0^{\infty} S^2\,p(S)\,dS = 10{,}600$$
$$\text{(Skew/crash-risk)} \qquad \mathbb{E}[\ln S_T] = \int_0^{\infty} \ln(S)\,p(S)\,dS = 4.60$$

---

## 1. Lagrangian Formulation

Introduce Lagrange multipliers $\lambda_0$ (normalization), $\lambda_1$ (mean), $\lambda_2$ (second moment), $\lambda_3$ (log-moment). Form the functional:

$$\mathcal{L}[p] = -\int_0^\infty p\ln p\,dS - \lambda_0\left(\int p\,dS - 1\right) - \lambda_1\left(\int Sp\,dS - 102\right) - \lambda_2\left(\int S^2 p\,dS - 10600\right) - \lambda_3\left(\int \ln(S)\,p\,dS - 4.60\right)$$

---

## 2. Functional Derivative

Taking $\partial \mathcal{L}/\partial p = 0$ pointwise (as in the Gaussian case):

$$-\ln p(S) - 1 - \lambda_0 - \lambda_1 S - \lambda_2 S^2 - \lambda_3 \ln S = 0$$

Solving for $p(S)$:

$$\boxed{p(S) = A\, S^{\lambda_3}\, \exp\left(-\lambda_1 S - \lambda_2 S^2\right), \qquad S > 0}$$

where $A = e^{-(1+\lambda_0)}$ absorbs the normalization constant. This is the **general functional form** of the maximum-entropy solution: an exponential-quadratic in $S$, multiplied by a power-law $S^{\lambda_3}$ coming from the log-moment constraint.

**Interpretation of each term:**
- The $e^{-\lambda_1 S}$ factor is the "exponential-family" response to the mean constraint (same role as in the standard exponential distribution).
- The $e^{-\lambda_2 S^2}$ factor is the Gaussian-like response to the second-moment/variance constraint — it must have $\lambda_2 > 0$ for the integral to converge as $S\to\infty$.
- The $S^{\lambda_3}$ factor is the response to the *log*-moment constraint, and it is what gives the distribution **skewness / fat-tail behavior near $S=0$** that a pure Gaussian cannot capture. This term is precisely why MaxEnt with this constraint set produces something richer than the Normal distribution used in Black–Scholes.

This density is a member of the **Generalized Inverse Gaussian–type / generalized Gamma-Normal family**, sometimes called a *generalized hyperbolic-type distribution restricted to the positive half-line*.

---

## 3. Solving for the Normalization Constant (Closed Form)

Complete the square in the exponent:

$$-\lambda_2 S^2 - \lambda_1 S = -\lambda_2\left(S + \frac{\lambda_1}{2\lambda_2}\right)^2 + \frac{\lambda_1^2}{4\lambda_2}$$

The normalizing integral is:

$$\frac{1}{A} = \int_0^{\infty} S^{\lambda_3} \exp\left(-\lambda_1 S - \lambda_2 S^2\right) dS$$

This integral does **not** have an elementary closed form (unlike the pure-Gaussian case in the Mean/Variance problem) because of the extra $S^{\lambda_3}$ power-law factor combined with the lower limit at $S=0$ rather than $-\infty$. It is expressible in terms of the **parabolic cylinder function** $D_\nu(\cdot)$ (equivalently, the confluent hypergeometric function $U(a,b,z)$):

$$\int_0^{\infty} S^{\nu-1} e^{-\lambda_1 S - \lambda_2 S^2}\,dS = (2\lambda_2)^{-\nu/2}\,\Gamma(\nu)\, e^{\lambda_1^2/(8\lambda_2)}\, D_{-\nu}\!\left(\frac{\lambda_1}{\sqrt{2\lambda_2}}\right), \qquad \nu = \lambda_3+1$$

So the exact normalization constant is:

$$A^{-1} = (2\lambda_2)^{-(\lambda_3+1)/2}\,\Gamma(\lambda_3+1)\, e^{\lambda_1^2/(8\lambda_2)}\, D_{-(\lambda_3+1)}\!\left(\frac{\lambda_1}{\sqrt{2\lambda_2}}\right)$$

---

## 4. Determining the Multipliers $(\lambda_1,\lambda_2,\lambda_3)$

The three multipliers must satisfy the moment-matching system:

$$\mathbb{E}[S] = 102, \qquad \mathbb{E}[S^2] = 10{,}600, \qquad \mathbb{E}[\ln S] = 4.60$$

where each expectation is computed under $p(S) = A\,S^{\lambda_3}e^{-\lambda_1 S - \lambda_2 S^2}$. Because $\ln A$ is the **cumulant generating function** of this exponential family, each moment constraint is equivalent to:

$$\mathbb{E}[S] = -\frac{\partial \ln Z}{\partial \lambda_1}, \qquad \mathbb{E}[S^2] = -\frac{\partial \ln Z}{\partial \lambda_2}, \qquad \mathbb{E}[\ln S] = -\frac{\partial \ln Z}{\partial \lambda_3}$$

with $Z(\lambda_1,\lambda_2,\lambda_3) \equiv A^{-1}$ from Section 3.

**This is a system of three coupled transcendental equations in $(\lambda_1,\lambda_2,\lambda_3)$ — there is no elementary analytic solution**, because $Z$ involves the parabolic cylinder function. This is expected: as soon as a MaxEnt problem includes more than the first two moments (mean and variance) simultaneously, the resulting normalizing constant generically requires special functions, and the multipliers must be found **numerically**, typically via:

- Newton–Raphson / gradient descent on the moment-matching residuals, or
- Convex optimization on the dual problem (since $\ln Z(\lambda)$ is convex in $\lambda$, this dual is well-behaved and numerically stable).

---

## 5. Sanity Check: Consistency of the Data

Before solving numerically, check the constraints are mutually consistent. The **variance** implied by the constraints is:

$$\text{Var}(S_T) = \mathbb{E}[S_T^2] - (\mathbb{E}[S_T])^2 = 10{,}600 - (102)^2 = 10{,}600 - 10{,}404 = 196$$

so the implied standard deviation is $\sigma = \sqrt{196} = 14$, i.e. about **13.7% monthly volatility** relative to the $102 forward price — a plausible, moderately volatile setting for a tech stock, consistent with the problem statement.

Compare this to what a **pure Gaussian** with the same mean and variance would imply for $\mathbb{E}[\ln S]$. Using a second-order (Jensen) approximation for a Gaussian $S \sim \mathcal{N}(102, 196)$:

$$\mathbb{E}[\ln S] \approx \ln(102) - \frac{\sigma^2}{2\mu^2} = \ln(102) - \frac{196}{2(102)^2} \approx 4.625 - 0.0094 \approx 4.616$$

The market-implied constraint is $\mathbb{E}[\ln S_T] = 4.60$, which is **lower** than the Gaussian-implied value of $\approx 4.616$. Since $\ln(\cdot)$ is concave, a lower $\mathbb{E}[\ln S]$ than the Gaussian benchmark indicates **extra downside mass / left-skew** relative to a Normal distribution — exactly the "fat left tail from put buying" described in the problem. This confirms $\lambda_3 \ne 0$ is required (i.e., a pure Gaussian is *not* consistent with all three constraints simultaneously), which is precisely why the MaxEnt solution departs from Black–Scholes' log-normal/Gaussian assumption.

---

## 6. Final Answer

$$\boxed{p(S_T) = A\, S_T^{\lambda_3}\, \exp\left(-\lambda_1 S_T - \lambda_2 S_T^2\right), \qquad S_T > 0}$$

with $(A,\lambda_1,\lambda_2,\lambda_3)$ determined by solving the four coupled equations (normalization + three moment constraints) numerically. This distribution:

- Reduces to a **truncated/half Gaussian** if $\lambda_3 = 0$ (i.e., if the log-moment constraint were dropped or happened to be Gaussian-consistent);
- Otherwise exhibits **skewness controlled by $\lambda_3$**, giving the model the fat-tailed, asymmetric downside risk that real option markets price in, without arbitrarily assuming a parametric family (e.g., log-normal) in advance — this is the core value-add of the Maximum Entropy approach over Black–Scholes.
