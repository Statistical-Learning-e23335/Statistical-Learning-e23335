# Q1) The Probability of Shooting at a Target

A point is chosen at random from a disk of radius $10$. Let $A$ be the event that the point lies within $1$ unit of the boundary.

1. Model the experiment assuming the point is uniformly distributed with respect to area, and compute $P(A)$.
2. Model the experiment assuming the distance to the point from the center is chosen uniformly from $[0,10]$, and the direction is chosen independently and uniformly from $[0,2\pi)$ and compute $P(A)$.
3. Explain why the two answers differ.


1. Uniform Distribution with Respect to AreaWhen a point is chosen uniformly with respect to area, the probability is the ratio of the favorable area to the total area.Total Area ($S$): A disk of radius $r = 10$.$$Area_{total} = \pi r^2 = \pi(10)^2 = 100\pi$$Event $A$: The point lies within 1 unit of the boundary. This means the point is in the outer ring (annulus) between radius $r = 9$ and $r = 10$.Favorable Area: Total area minus the area of the inner disk (radius 9).$$Area_{A} = \pi(10)^2 - \pi(9)^2 = 100\pi - 81\pi = 19\pi$$Probability $P(A)$:$$P(A) = \frac{19\pi}{100\pi} = \mathbf{0.19}$$2. Uniform Distribution with Respect to DistanceIn this model, we choose a radius $R$ uniformly from $[0, 10]$ and an angle independently. The angle doesn't affect the event $A$, as $A$ only depends on the distance from the center.Variable $R$: $R \sim \text{Uniform}(0, 10)$.Event $A$: The point is within 1 unit of the boundary, which means $9 \le R \le 10$.Probability $P(A)$: Since the distribution of $R$ is uniform over the length of the interval $[0, 10]$:$$P(A) = P(9 \le R \le 10) = \frac{10 - 9}{10 - 0} = \frac{1}{10} = \mathbf{0.1}$$3. Why do the answers differ?The difference lies in how "density" is distributed across the disk.In the first case (Area), the probability depends on where the physical space is. Because there is significantly more area near the outer edge of a circle than near the center, a point chosen uniformly by area is much more likely to land near the boundary.In the second case (Distance), you are essentially picking a point on a line segment from 0 to 10 and then "spinning" it around. This treats the distance near the center as being just as likely as the distance near the edge.

# Q2) Bertrand's Paradox

A chord of a circle is chosen “at random.” What is the probability that its length exceeds the side length of the inscribed equilateral triangle?

Show that different mathematically natural models of “random chord” lead to different answers.

1. The "Random Endpoints" MethodIn this model, we choose two points on the circumference of the circle independently and uniformly at random and connect them to form a chord.The Logic: Fix one point as a vertex of the inscribed equilateral triangle. The other endpoint must fall on the arc between the other two vertices of the triangle for the chord to be longer than the side length.The Calculation: The triangle divides the circumference into three equal arcs, each spanning $120^\circ$. Only the middle arc results in a chord longer than the side.Result: $P = \frac{1}{3}$
2. The "Random Radial Distance" MethodIn this model, we choose a radius of the circle and then choose a point on that radius uniformly at random. We construct the chord perpendicular to the radius at that point.The Logic: As noted earlier, a chord is longer than the side of the inscribed triangle if its distance from the center is less than $R/2$.The Calculation: Since we are picking a point on a radius of length $R$, the favorable region is the segment from the center to $R/2$.Result: $P = \frac{R/2}{R} = \frac{1}{2}$
3. The "Random Midpoint" MethodIn this model, we choose a point anywhere inside the circle uniformly at random and treat it as the midpoint of a chord.The Logic: A chord's length is uniquely determined by its midpoint. For the chord to be longer than the triangle's side, its midpoint must fall within a smaller, concentric circle with radius $R/2$.The Calculation: We compare the area of the smaller circle to the area of the larger circle.$Area_{total} = \pi R^2$$Area_{favorable} = \pi (R/2)^2 = \frac{1}{4}\pi R^2$Result: $P = \frac{\pi (R/2)^2}{\pi R^2} = \frac{1}{4}$

#Q8) Probability of Being a Girl?

A family has two children. Assume each child is equally likely to be a boy or a girl, independently of the other.

What is the probability that both children are girls, given that **at least one** of them is a girl?

1. Identify the Full Sample Space    
When a family has two children, there are four equally likely outcomes based on the order of birth (Boy-Girl is distinct from Girl-Boy):BB (Boy, then Boy)BG (Boy, then Girl)GB (Girl, then Boy)GG (Girl, then Girl)
2. Apply the ConditionThe problem provides a specific piece of information: at least one of them is a girl.       
This "given" information allows us to eliminate any outcomes that don't fit. Looking at our list above, the outcome BB is now impossible. Our reduced sample space is:BGGBGGThere are now only 3 possible equally likely outcomes.
3. Calculate the ProbabilityWe want to find the probability that both children are girls (GG) within this new, restricted set of possibilities.   
Number of favorable outcomes: 1 (which is GG)Total number of possible outcomes: 3 (which are BG, GB, GG)$$P(\text{Both Girls} \mid \text{At least one Girl}) = \frac{1}{3}$$

# Q9) Information Available in a Discrete Random Variable

A factory inspects a product and records its quality using two different coding systems.

Each product falls into exactly one of the following four categories:

* $G_1$: good quality, produced on Machine 1
* $G_2$: good quality, produced on Machine 2
* $D_1$: defective, produced on Machine 1
* $D_2$: defective, produced on Machine 2

Assume the sample space is

$$
\Omega=\{G_1,G_2,D_1,D_2\},
$$

with probabilities

$$
P(\{G_1\})=0.50,\qquad P(\{G_2\})=0.20,\qquad P(\{D_1\})=0.10,\qquad P(\{D_2\})=0.20.
$$

Define two discrete random variables on the same probability space:

* $X$ records only whether the item is good or defective:
  $$
  X(G_1)=0,\quad X(G_2)=0,\quad X(D_1)=1,\quad X(D_2)=1.
  $$

* $Y$ records both quality and machine:
  $$
  Y(G_1)=1,\quad Y(G_2)=2,\quad Y(D_1)=3,\quad Y(D_2)=4.
  $$

Answer the following.

1. Write down the probability space $(\Omega,\mathfrak F,P)$, where $\mathfrak F=\mathfrak P(\Omega)$.

2. Compute the $\sigma$-algebra generated by $X$, namely $\sigma(X)$.

3. Compute the $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$.

4. Show that $\sigma(X)\subseteq \sigma(Y)$, and explain why $Y$ contains more information than $X$.

5. Find the marginal distribution of $X$.

6. Find the marginal distribution of $Y$.

7. Compute the joint probabilities
   $$
   P(X=0,Y=1),\quad P(X=0,Y=2),\quad P(X=1,Y=3),\quad P(X=1,Y=4).
   $$

8. Compute the conditional probabilities
   $$
   P(X=1\mid Y=3),\qquad P(X=1\mid Y=4),\qquad P(Y=3\mid X=1),\qquad P(Y=4\mid X=1).
   $$

9. Interpret the difference between conditioning on $X$ and conditioning on $Y$.

10. Explain, in words, what information is available in $\sigma(X)$ but not in $\sigma(Y)$, and what information is available in $\sigma(Y)$ but not in $\sigma(X)$.

11. Give a Python based visualization of the model.

12. Find the marginal, joint, and conditional Shannon Entropy of $X$ and $Y$.

This problem explores the relationship between random variables, the information they encode, and the mathematical structures (sigma-algebras) that represent that information.
1. The Probability Space $(\Omega, \mathfrak{F}, P)$Sample Space: $\Omega = \{G_1, G_2, D_1, D_2\}$.Sigma-algebra: $\mathfrak{F} = \mathfrak{P}(\Omega)$ is the power set of $\Omega$, containing all $2^4 = 16$ possible subsets of outcomes.Probability Measure: Defined by the given weights: $P(G_1)=0.5, P(G_2)=0.2, P(D_1)=0.1, P(D_2)=0.2$.
2. Sigma-algebra Generated by $X$, $\sigma(X)$The sigma-algebra $\sigma(X)$ contains all sets that can be described using only the values of $X$.$X=0$ corresponds to $\{G_1, G_2\}$ (Good).$X=1$ corresponds to $\{D_1, D_2\}$ (Defective).$\sigma(X) = \{\emptyset, \{G_1, G_2\}, \{D_1, D_2\}, \Omega\}$.
3. Sigma-algebra Generated by $Y$, $\sigma(Y)$Since $Y$ assigns a unique value to every individual outcome in $\Omega$, it generates the finest possible partition.$\sigma(Y) = \mathfrak{P}(\Omega)$ (The full power set, containing all 16 subsets).
4. Inclusion and InformationProof: Every set in $\sigma(X)$ is also in $\sigma(Y)$ (e.g., $\{G_1, G_2\} = \{G_1\} \cup \{G_2\}$). Thus, $\sigma(X) \subseteq \sigma(Y)$.Explanation: $Y$ contains more information because it distinguishes between the machines. Knowing $X=0$ only tells you the item is good; knowing $Y$ tells you it is good and which machine made it.
5. Marginal Distribution of $X$$P(X=0) = P(\{G_1, G_2\}) = 0.50 + 0.20 = \mathbf{0.70}$$P(X=1) = P(\{D_1, D_2\}) = 0.10 + 0.20 = \mathbf{0.30}$
6. Marginal Distribution of $Y$$P(Y=1) = \mathbf{0.50}$$P(Y=2) = \mathbf{0.20}$$P(Y=3) = \mathbf{0.10}$$P(Y=4) = \mathbf{0.20}$
7. Joint Probabilities$P(X=0, Y=1) = P(\{G_1\}) = \mathbf{0.50}$$P(X=0, Y=2) = P(\{G_2\}) = \mathbf{0.20}$$P(X=1, Y=3) = P(\{D_1\}) = \mathbf{0.10}$$P(X=1, Y=4) = P(\{D_2\}) = \mathbf{0.20}$
8. Conditional Probabilities$P(X=1 \mid Y=3) = \frac{P(X=1, Y=3)}{P(Y=3)} = \frac{0.1}{0.1} = \mathbf{1.0}$$P(X=1 \mid Y=4) = \frac{P(X=1, Y=4)}{P(Y=4)} = \frac{0.2}{0.2} = \mathbf{1.0}$$P(Y=3 \mid X=1) = \frac{P(X=1, Y=3)}{P(X=1)} = \frac{0.1}{0.3} = \mathbf{1/3}$$P(Y=4 \mid X=1) = \frac{P(X=1, Y=4)}{P(X=1)} = \frac{0.2}{0.3} = \mathbf{2/3}$
9. Interpretation of ConditioningConditioning on $Y$ removes all uncertainty about $X$ (as seen by the probabilities of 1.0), because $Y$ is a "finer" variable. Conditioning on $X$ only reduces uncertainty; if we know an item is defective ($X=1$), we still have uncertainty about whether it came from Machine 1 ($1/3$) or Machine 2 ($2/3$).
10. Information ComparisonIn $\sigma(Y)$ but not $\sigma(X)$: The specific identity of the machine (Machine 1 vs. Machine 2).In $\sigma(X)$ but not $\sigma(Y)$: Nothing. Because $\sigma(X) \subseteq \sigma(Y)$, every piece of information $X$ provides is already inherently known if you know $Y$.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data Setup
outcomes = ['G1 (Y=1)', 'G2 (Y=2)', 'D1 (Y=3)', 'D2 (Y=4)']
probs = [0.5, 0.2, 0.1, 0.2]
colors = ['#4CAF50', '#4CAF50', '#F44336', '#F44336'] # Green for Good, Red for Defective

plt.figure(figsize=(10, 6))
bars = plt.bar(outcomes, probs, color=colors)

plt.title('Probability Distribution of Machine/Quality Outcomes (Y)')
plt.ylabel('Probability')
plt.ylim(0, 0.6)

# Add X labels
plt.text(0.5, -0.08, 'X = 0 (Good Items)', ha='center', fontweight='bold')
plt.text(2.5, -0.08, 'X = 1 (Defective Items)', ha='center', fontweight='bold')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01, f'{yval:.2f}', ha='center')

plt.show()
```

# Q10) Information Available in Continuous Random Varaibles

A chemical plant monitors each production run using two different measurement systems.

For a given run, let

* $T$ be the deviation of the reactor temperature from target, measured in standardized units,
* $P$ be the deviation of the reactor pressure from target, also measured in standardized units.

Assume that $(T,P)$ is jointly continuous, with independent normal components:

$$
T \sim N(0,1), \qquad P \sim N(0,4),
$$

and $T$ and $P$ are independent.

Let the underlying probability space be

$$
\Omega=\mathbb R^2, \qquad \mathfrak F=\mathfrak B(\mathbb R^2),
$$

and let $P$ be the probability measure induced by the joint density

$$
f_{T,P}(t,p)=\frac{1}{4\pi}\exp\left(-\frac{t^2}{2}-\frac{p^2}{8}\right), \qquad (t,p)\in\mathbb R^2.
$$

Define two random variables on the same probability space:

* $X:\Omega\to\mathbb R$ by
  $$
  X(t,p)=t,
  $$
  so that $X$ records only the reactor temperature deviation;

* $Y:\Omega\to\mathbb R^2$ by
  $$
  Y(t,p)=(t,p),
  $$
  so that $Y$ records the full sensor state: both temperature and pressure deviations.

Answer the following.

1. Write down the probability space $(\Omega,\mathfrak F,P)$ explicitly.

2. Compute the $\sigma$-algebra generated by $X$, namely $\sigma(X)$.

3. Compute the $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$.

4. Show that
   $$
   \sigma(X)\subseteq \sigma(Y),
   $$
   and explain why $Y$ contains more information than $X$.

5. Find the marginal density of $X$.

6. Find the marginal density of $Y$.

7. Compute the following probabilities:
   $$
   P(X\leq 0), \qquad P(X>1), \qquad P(Y\in (-\infty,0]\times\mathbb R), \qquad P(Y\in [ -1,1]\times[-2,2]).
   $$

8. Compute the conditional distribution of $X$ given $Y=(t,p)$.

9. Compute the conditional distribution of $Y$ given $X=t$.

10. Explain, in words, what is known when one conditions on $X$ and what additional information becomes available when one conditions on $Y$.

11. Show that
    $$
    X=\pi_1\circ Y,
    $$
    where $\pi_1(t,p)=t$, and explain why this implies that $X$ is a measurable function of $Y$.

12. Give a Python-based visualization of the model, for example by plotting:

* the joint density of $Y=(T,P)$ as a contour plot or heat map;
* the marginal density of $X$;
* and a diagram illustrating that $\sigma(X)$ is coarser than $\sigma(Y)$.

13. Compute the differential entropy of $X$:

$$
h(X)=-\int_{\mathbb R} f_X(x)\ln f_X(x)\,dx.
$$

14. Compute the differential entropy of $Y$:

$$
h(Y)=-\int_{\mathbb R^2} f_Y(t,p)\ln f_Y(t,p)\,dt\,dp.
$$

15. Compute the conditional differential entropy

$$
h(Y|X),
$$

and interpret it physically.

16. Discuss what happens to

$$
h(X|Y)
$$

in this model, and explain why the continuous case differs from the discrete case when one random variable is completely determined by the other.


1. The Probability Space $(\Omega, \mathfrak{F}, P)$Sample Space: $\Omega = \mathbb{R}^2$ (representing all possible pairs of temperature and pressure deviations).Sigma-algebra: $\mathfrak{F} = \mathfrak{B}(\mathbb{R}^2)$, the Borel $\sigma$-algebra on $\mathbb{R}^2$.Probability Measure: For any set $B \in \mathfrak{F}$:$$P(B) = \int_B \frac{1}{4\pi}\exp\left(-\frac{t^2}{2}-\frac{p^2}{8}\right) dt\,dp$$
2. Sigma-algebra Generated by $X$, $\sigma(X)$$X$ only "sees" the temperature.$\sigma(X) = \{ B \times \mathbb{R} : B \in \mathfrak{B}(\mathbb{R}) \}$.This represents "vertical strips" in the plane. You can determine if the temperature is in a certain range, but you know nothing about pressure.
3. Sigma-algebra Generated by $Y$, $\sigma(Y)$Since $Y(t,p) = (t,p)$ is the identity map, it preserves all information.$\sigma(Y) = \mathfrak{B}(\mathbb{R}^2)$ (the entire Borel $\sigma$-algebra).
4. Inclusion and InformationProof: Every set in $\sigma(X)$ (a vertical strip) is a Borel set in $\mathbb{R}^2$, so $\sigma(X) \subseteq \sigma(Y)$.Explanation: $Y$ contains more information because it specifies a precise point $(t, p)$ in the plane, whereas $X$ only specifies a coordinate line $t$. Knowing $Y$ tells you both reactor metrics; knowing $X$ leaves the pressure completely unknown.
5 & 6. Marginal DensitiesMarginal $X$: $f_X(t) = \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{t^2}{2}\right)$ (Standard Normal $N(0,1)$).Marginal $Y$: $f_Y(t,p) = f_{T,P}(t,p) = \frac{1}{\sqrt{2\pi}(1)} e^{-t^2/2} \cdot \frac{1}{\sqrt{2\pi}(2)} e^{-p^2/8}$.
7. Probabilities$P(X \leq 0) = \mathbf{0.5}$ (Symmetry of the normal distribution).$P(X > 1) \approx \mathbf{0.1587}$ (From standard Z-table).$P(Y \in (-\infty,0] \times \mathbb{R}) = P(X \leq 0) = \mathbf{0.5}$.$P(Y \in [-1,1] \times [-2,2]) = P(-1 \leq T \leq 1) \cdot P(-2 \leq P \leq 2)$.$P(-1 \leq T \leq 1) \approx 0.6827$.$P(-1 \leq P/2 \leq 1) \approx 0.6827$.Total $\approx 0.6827^2 \approx \mathbf{0.466}$.

8 & 9. Conditional Distributions$X|Y=(t,p)$: This is a degenerate distribution (a Dirac delta $\delta_t$). Since $Y$ includes $T$, $X$ is known exactly.$Y|X=t$: Since $T$ and $P$ are independent, knowing $T=t$ tells us nothing about $P$. The distribution is a vertical line at $t$ where the pressure $P$ remains $N(0,4)$.

10. Information InterpretationConditioning on $X$ provides the "horizontal" location (temperature). Conditioning on $Y$ provides the "vertical" location (pressure), removing all remaining uncertainty about the reactor's state.
11. Measurability$X = \pi_1 \circ Y$. Because the projection $\pi_1$ is a continuous function, and $Y$ is the identity, $X$ is a measurable function of $Y$. This confirms mathematically that $Y$ "contains" $X$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, multivariate_normal

# 1. Joint Density (Y)
x, y = np.mgrid[-3:3:.01, -6:6:.01]
pos = np.dstack((x, y))
rv = multivariate_normal([0, 0], [[1, 0], [0, 4]])

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Heatmap for Y
ax[0].contourf(x, y, rv.pdf(pos), cmap='viridis')
ax[0].set_title('Joint Density f(T,P) - sigma(Y)')
ax[0].set_xlabel('Temperature (T)')
ax[0].set_ylabel('Pressure (P)')

# 2. Marginal Density (X)
t_axis = np.linspace(-4, 4, 100)
ax[1].plot(t_axis, norm.pdf(t_axis, 0, 1), color='red', lw=2)
ax[1].fill_between(t_axis, norm.pdf(t_axis, 0, 1), alpha=0.3, color='red')
ax[1].set_title('Marginal Density f(X)')
ax[1].set_xlabel('T')

plt.tight_layout()
plt.show()
```

Differential EntropyThe entropy of a Normal $N(\mu, \sigma^2)$ is $h = \ln(\sigma\sqrt{2\pi e})$.$h(X)$: $\ln(1\sqrt{2\pi e}) \approx \mathbf{1.419}$ nats.$h(Y)$: Since $T, P$ are independent, $h(Y) = h(T) + h(P)$.$h(P) = \ln(2\sqrt{2\pi e})$.$h(Y) = 1.419 + 2.112 \approx \mathbf{3.531}$ nats.$h(Y|X)$: $h(Y) - h(X) = h(P) \approx \mathbf{2.112}$ nats.Physical Interpretation: This is the residual uncertainty in the reactor state (the pressure) that remains after you have already measured the temperature.

16. The $h(X|Y)$ ParadoxIn the discrete model (image_64a7a4.png), $H(X|Y)=0$. In the continuous case, if $X$ is a function of $Y$, $h(X|Y)$ technically goes to $-\infty$.Why? In continuous space, a single point has zero width. If you know $Y$ exactly, the "volume" of uncertainty for $X$ is zero. Since $\ln(0) = -\infty$, differential entropy reflects that a perfectly known continuous variable contains "infinite" information relative to a spread-out one.
