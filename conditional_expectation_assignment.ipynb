#Q1) Prediction of fraudulent transactions

A bank monitors credit-card transactions for possible fraud.

For each transaction, let the sample space be

$$
\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\},
$$

where:

* $\omega_1$: low-risk customer, legitimate transaction,
* $\omega_2$: low-risk customer, fraudulent transaction,
* $\omega_3$: high-risk customer, legitimate transaction,
* $\omega_4$: high-risk customer, fraudulent transaction.

Assume the probabilities are

$$
P(\{\omega_1\})=0.50,\qquad
P(\{\omega_2\})=0.10,\qquad
P(\{\omega_3\})=0.20,\qquad
P(\{\omega_4\})=0.20.
$$

Let $X:\Omega\to\mathbb R$ denote the fraud indicator:

$$
X(\omega_1)=0,\qquad
X(\omega_2)=1,\qquad
X(\omega_3)=0,\qquad
X(\omega_4)=1.
$$

Thus $X=1$ means the transaction is fraudulent, and $X=0$ means it is legitimate.

Now suppose the bank does not observe the full outcome $\omega$, but only the customer's risk class. Define $Y:\Omega\to\mathbb R$ by

$$
Y(\omega_1)=0,\qquad
Y(\omega_2)=0,\qquad
Y(\omega_3)=1,\qquad
Y(\omega_4)=1.
$$

Thus:

* $Y=0$ means “low-risk customer,”
* $Y=1$ means “high-risk customer.”

Answer the following.

1. Write down the probability space $(\Omega,\mathfrak F,P)$, where $\mathfrak F=\mathfrak P(\Omega)$.

2. Compute the $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$.

3. Explain why $E[X\mid Y]$ must be $\sigma(Y)$-measurable, and therefore must be constant on the sets $\{ \omega_1,\omega_2\}$ and $\{\omega_3,\omega_4\}$.

4. Compute $E[X\mid Y]$ explicitly and show that

$$
E[X\mid Y]= \frac{1}{6} \mathbb{1}_{\{Y=0\}} + \frac{1}{2} \mathbb{1}_{\{Y=1\}}
$$

5. Verify the defining property of conditional expectation by checking that for every set $A\in\sigma(Y)$,

$$
\int_A E[X\mid Y]\,dP=\int_A X\,dP.
$$

6. Find the Conditional PMF $p_{X|Y}(x \mid y)$.

7. Verify the identity:
$$\mathbb{E}[X\mid Y=y] = \sum_{x \in \{0,1\}} x \cdot p_{X|Y}(x \mid y)$$.

8. Calculate the **Mean Square Error (MSE)**

9. How would the **Mean Square Error (MSE)** change if the bank decided to ignore the customer risk class $Y$ and instead used the global average fraud rate ($E[X] = 0.30$) as a constant estimate for every transaction? Calculate this error and explain its relationship to the **Law of Total Variance**.

10. Using the joint distribution of $(X, Y)$ derived from the sample space:

    a) Compute the entropy $H(X)$, representing the bank's initial uncertainty regarding fraud.

    b) Compute the conditional entropy $H(X \mid Y)$, representing the uncertainty remaining after the customer's risk class is revealed.

    c) Calculate the Mutual Information $I(X; Y)$.

    d) Based on these values, what percentage of the total uncertainty about fraud is removed by knowing the customer's risk class?

1. Mathematical Structure of the Probability Space $(\Omega, \mathcal{F}, P)$A foundational probability space is defined by the triplet $(\Omega, \mathcal{F}, P)$. Given that the $\sigma$-algebra $\mathcal{F}$ is the power set $\mathcal{P}(\Omega)$, the space is structured as follows:Sample Space ($\Omega$):$$\Omega = \{\omega_1, \omega_2, \omega_3, \omega_4\}$$$\sigma$-algebra ($\mathcal{F}$): Since $|\Omega| = 4$, the power set contains $2^4 = 16$ elements:$$\mathcal{F} = \big\{ \emptyset, \{\omega_1\}, \{\omega_2\}, \{\omega_3\}, \{\omega_4\}, \{\omega_1, \omega_2\}, \{\omega_1, \omega_3\}, \{\omega_1, \omega_4\}, \{\omega_2, \omega_3\}, \{\omega_2, \omega_4\}, \{\omega_3, \omega_4\}, \{\omega_1, \omega_2, \omega_3\}, \{\omega_1, \omega_2, \omega_4\}, \{\omega_1, \omega_3, \omega_4\}, \{\omega_2, \omega_3, \omega_4\}, \Omega \big\}$$Probability Measure ($P$): A mapping $P: \mathcal{F} \to [0, 1]$ uniquely determined by its values on the singletons (atoms):$$P(\{\omega_1\}) = 0.50, \quad P(\{\omega_2\}) = 0.10, \quad P(\{\omega_3\}) = 0.20, \quad P(\{\omega_4\}) = 0.20$$For any compound event $A \in \mathcal{F}$, the measure satisfies countable additivity: $P(A) = \sum_{\omega \in A} P(\{\omega\})$.
2. Derivation of the Generated $\sigma$-algebra $\sigma(Y)$The random variable $Y: \Omega \to \mathbb{R}$ acts as a measurability constraint. The $\sigma$-algebra generated by $Y$ is defined as the smallest sub-$\sigma$-algebra of $\mathcal{F}$ with respect to which $Y$ is measurable. This is given by the collection of all pre-images of Borel sets under $Y$:$$\sigma(Y) = \{ Y^{-1}(B) : B \in \mathcal{B}(\mathbb{R}) \}$$Evaluating the pre-images for the range of $Y$:$Y^{-1}(\{0\}) = \{\omega \in \Omega : Y(\omega) = 0\} = \{\omega_1, \omega_2\}$$Y^{-1}(\{1\}) = \{\omega \in \Omega : Y(\omega) = 1\} = \{\omega_3, \omega_4\}$Taking all possible structural unions, intersections, and complements of these fundamental pre-images yields:$$\sigma(Y) = \big\{ \emptyset, \{\omega_1, \omega_2\}, \{\omega_3, \omega_4\}, \Omega \big\}$$
3. Measurability Requirements of $\mathbb{E}[X \mid Y]$By definition, the conditional expectation $\mathbb{E}[X \mid Y]$ (alternatively written as $\mathbb{E}[X \mid \sigma(Y)]$) is a random variable mapping $\Omega \to \mathbb{R}$ that must satisfy $\sigma(Y)$-measurability.For a random variable to be measurable with respect to a specific $\sigma$-algebra, its level sets must be elements of that $\sigma$-algebra. The atoms (minimal non-empty evaluation blocks) of $\sigma(Y)$ are $A_0 = \{\omega_1, \omega_2\}$ and $A_1 = \{\omega_3, \omega_4\}$.If $\mathbb{E}[X \mid Y]$ took distinct values within an atom—for instance, if $\mathbb{E}[X \mid Y](\omega_1) \neq \mathbb{E}[X \mid Y](\omega_2)$—then the pre-image of one of those individual values would yield the singleton $\{\omega_1\}$ or $\{\omega_2\}$. Because neither $\{\omega_1\}$ nor $\{\omega_2\}$ belongs to $\sigma(Y)$, such a function would violate the measurability requirement. Hence, $\mathbb{E}[X \mid Y]$ must be constant across each atom of the conditioning $\sigma$-algebra.
4. Explicit Computation of the Radon-Nikodym Derivative $\mathbb{E}[X \mid Y]$Because $\sigma(Y)$ is generated by a countable, disjoint partition of $\Omega$, the conditional expectation can be computed directly using the standard sub-space expectation formula over each partition block:On the atom $A_0 = \{Y = 0\} = \{\omega_1, \omega_2\}$:$$\mathbb{E}[X \mid Y = 0] = \frac{\int_{A_0} X \, dP}{P(A_0)} = \frac{X(\omega_1)P(\{\omega_1\}) + X(\omega_2)P(\{\omega_2\})}{P(\{\omega_1\}) + P(\{\omega_2\})}$$$$\mathbb{E}[X \mid Y = 0] = \frac{(0)(0.50) + (1)(0.10)}{0.50 + 0.10} = \frac{0.10}{0.60} = \frac{1}{6}$$On the atom $A_1 = \{Y = 1\} = \{\omega_3, \omega_4\}$:$$\mathbb{E}[X \mid Y = 1] = \frac{\int_{A_1} X \, dP}{P(A_1)} = \frac{X(\omega_3)P(\{\omega_3\}) + X(\omega_4)P(\{\omega_4\})}{P(\{\omega_3\}) + P(\{\omega_4\})}$$$$\mathbb{E}[X \mid Y = 1] = \frac{(0)(0.20) + (1)(0.20)}{0.20 + 0.20} = \frac{0.20}{0.40} = \frac{1}{2}$$Expressing this as a functional mapping across the full sample space using indicator functions:$$\mathbb{E}[X \mid Y] = \frac{1}{6}\mathbb{1}_{\{\omega_1, \omega_2\}} + \frac{1}{2}\mathbb{1}_{\{\omega_3, \omega_4\}} = \frac{1}{6}\mathbb{1}_{\{Y=0\}} + \frac{1}{2}\mathbb{1}_{\{Y=1\}}$$
5. Verification of the Defining Conditional PropertyThe conditional expectation must satisfy the defining functional equation for every $A \in \sigma(Y)$:$$\int_A \mathbb{E}[X \mid Y] \, dP = \int_A X \, dP
$$We verify this exhaustively across all four elements of $\sigma(Y)$:Case 1: $A = \emptyset$$\int_{\emptyset} \mathbb{E}[X \mid Y] \, dP = 0 = \int_{\emptyset} X \, dP \quad \checkmark$$Case 2: A = \{\omega_1, \omega_2\}$$\int_A \mathbb{E}[X \mid Y] \, dP = \sum_{\omega \in A} \mathbb{E}[X \mid Y](\omega) P(\{\omega\}) = \frac{1}{6}(0.50) + \frac{1}{6}(0.10) = \frac{1}{6}(0.60) = 0.10$$$$\int_A X \, dP = \sum_{\omega \in A} X(\omega) P(\{\omega\}) = 0(0.50) + 1(0.10) = 0.10 \quad \checkmark$$Case 3: $$A = \{\omega_3, \omega_4\}\int_A \mathbb{E}[X \mid Y] \, dP = \sum_{\omega \in A} \mathbb{E}[X \mid Y](\omega) P(\{\omega\}) = \frac{1}{2}(0.20) + \frac{1}{2}(0.20) = \frac{1}{2}(0.40) = 0.20$$$$\int_A X \, dP = \sum_{\omega \in A} X(\omega) P(\{\omega\}) = 0(0.20) + 1(0.20) = 0.20 \quad \checkmark$$Case 4: $$A = \Omega\int_{\Omega} \mathbb{E}[X \mid Y] \, dP = 0.10 + 0.20 = 0.30$$$$\int_{\Omega} X \, dP = \mathbb{E}[X] = 0(0.50) + 1(0.10) + 0(0.20) + 1(0.20) = 0.30 \quad \checkmark$$

6. Derivation of the Conditional Probability Mass Function $$p_{X \mid Y}(x \mid y)$$The conditional PMF is defined via the conditional probability formula $$P(X = x \mid Y = y) = \frac{P(X=x, Y=y)}{P(Y=y)}$$:Evaluating y = 0$:$p_{X \mid Y}(0 \mid 0) = P(\{\omega_1\} \mid \{\omega_1, \omega_2\}) = \frac{0.50}{0.60} = \frac{5}{6}$$p_{X \mid Y}(1 \mid 0) = P(\{\omega_2\} \mid \{\omega_1, \omega_2\}) = \frac{0.10}{0.60} = \frac{1}{6}$Evaluating $y = 1$:$p_{X \mid Y}(0 \mid 1) = P(\{\omega_3\} \mid \{\omega_3, \omega_4\}) = \frac{0.20}{0.40} = \frac{1}{2}$$p_{X \mid Y}(1 \mid 1) = P(\{\omega_4\} \mid \{\omega_3, \omega_4\}) = \frac{0.20}{0.40} = \frac{1}{2}$Structured Joint-to-Conditional PMF Matrix$ pX∣Y​(x∣y)x=0 (Legitimate)x=1 (Fraudulent)$y = 0$ (Low-risk)$\frac${5}{6}$$\frac{1}{6}$$y = 1$ (High-risk)$\frac{1}{2}$$\frac{1}{2}$

7. Verification of the Canonical Identity $\mathbb{E}[X \mid Y = y] = \sum_{x} x \cdot p_{X \mid Y}(x \mid y)$We verify that the measure-theoretic definition aligns perfectly with the standard undergraduate algebraic definition:Condition $y = 0$:$$\sum_{x \in \{0,1\}} x \cdot p_{X \mid Y}(x \mid 0) = 0 \cdot \left(\frac{5}{6}\right) + 1 \cdot \left(\frac{1}{6}\right) = \frac{1}{6} = \mathbb{E}[X \mid Y=0] \quad \checkmark$$Condition $y = 1$:$$\sum_{x \in \{0,1\}} x \cdot p_{X \mid Y}(x \mid 1) = 0 \cdot \left(\frac{1}{2}\right) + 1 \cdot \left(\frac{1}{2}\right) = \frac{1}{2} = \mathbb{E}[X \mid Y=1] \quad \checkmark$$
8. Calculation of the Minimum Mean Square Error (MMSE)The conditional expectation $\mathbb{E}[X \mid Y]$ acts as the orthogonal projection of $X$ onto the $L^2(\sigma(Y))$ sub-space, minimizing the Mean Square Error. Let $g(Y) = \mathbb{E}[X \mid Y]$:$$\text{MSE} = \mathbb{E}\left[(X - \mathbb{E}[X \mid Y])^2\right] = \sum_{\omega \in \Omega} (X(\omega) - g(Y(\omega)))^2 \cdot P(\{\omega\})$$Evaluating this term-by-term across the sample space:$\omega_1$: $\left(0 - \frac{1}{6}\right)^2 \cdot 0.50 = \frac{1}{36} \cdot \frac{1}{2} = \frac{1}{72}$$\omega_2$: $\left(1 - \frac{1}{6}\right)^2 \cdot 0.10 = \frac{25}{36} \cdot \frac{1}{10} = \frac{25}{360} = \frac{5}{72}$$\omega_3$: $\left(0 - \frac{1}{2}\right)^2 \cdot 0.20 = \frac{1}{4} \cdot \frac{1}{5} = \frac{1}{20}$$\omega_4$: $\left(1 - \frac{1}{2}\right)^2 \cdot 0.20 = \frac{1}{4} \cdot \frac{1}{5} = \frac{1}{20}$$\text{MSE} = \left(\frac{1}{72} + \frac{5}{72}\right) + \left(\frac{1}{20} + \frac{1}{20}\right) = \frac{6}{72} + \frac{2}{20} = \frac{1}{12} + \frac{1}{10} = \frac{11}{60} \approx 0.18333$$

9. Structural Analysis under Zero-Information Predictor & Law of Total VarianceIf the predictor ignores the conditioning filtration $\sigma(Y)$, the problem reduces to a constant optimization problem in $L^2(\Omega)$, where the minimizing value is the unconditional mean: $\mathbb{E}[X] = 0.30$.The performance error under this constraint is equivalent to the absolute variance of $X$. Because $X \in \{0, 1\}$, it is an indicator variable where $X^2 = X$, giving $\mathbb{E}[X^2] = \mathbb{E}[X] = 0.30$:$$\text{MSE}_{\text{constant}} = \text{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2 = 0.30 - (0.30)^2 = 0.21$$Deep-Dive into the Law of Total VarianceThe total variance can be decomposed into the unexplained and explained components:$$\text{Var}(X) = \mathbb{E}[\text{Var}(X \mid Y)] + \text{Var}(\mathbb{E}[X \mid Y])$$Unexplained Component ($\mathbb{E}[\text{Var}(X \mid Y)]$): This represents the residual risk that remains even after tracking $Y$. It matches our computed MMSE value:$$\mathbb{E}[\text{Var}(X \mid Y)] = \frac{11}{60} \approx 0.18333$$Explained Component ($\text{Var}(\mathbb{E}[X \mid Y])$): The reduction in error achieved by leveraging the information system $Y$:$$\text{Var}(\mathbb{E}[X \mid Y]) = \mathbb{E}\left[(\mathbb{E}[X \mid Y])^2\right] - (\mathbb{E}[X])^2 = \left[ \left(\frac{1}{6}\right)^2(0.60) + \left(\frac{1}{2}\right)^2(0.40) \right] - 0.09$$$$\text{Var}(\mathbb{E}[X \mid Y]) = \left[ \frac{1}{36} \cdot \frac{3}{5} + \frac{1}{4} \cdot \frac{2}{5} \right] - 0.09 = \frac{7}{60} - \frac{5.4}{60} = \frac{1.6}{60} = \frac{4}{150} \approx 0.02667$$$$\text{Total Variance} = 0.18333 + 0.02667 = 0.21$$Ignoring $Y$ forces the bank to absorb the variance of the predictor itself ($\approx 0.02667$), increasing the overall system error to $0.21$.
10. Information-Theoretic Calculationsa) Unconditional Entropy $H(X)$The raw uncertainty of the fraud variable before conditioning:$$H(X) = -\sum_{x \in \{0,1\}} p_X(x) \log_2 p_X(x) = -0.70 \log_2(0.70) - 0.30 \log_2(0.30)$$$$H(X) \approx -(0.70 \cdot -0.51457) - (0.30 \cdot -1.73697) = 0.36020 + 0.52109 = 0.88129 \text{ bits}$$b) Conditional Entropy $H(X \mid Y)$The expected residual uncertainty of $X$ after conditioning on the information field $Y$:$$H(X \mid Y) = P(Y=0)H(X \mid Y=0) + P(Y=1)H(X \mid Y=1)$$Sub-entropy $H(X \mid Y=0)$:$$H(X \mid Y=0) = -\frac{5}{6}\log_2\left(\frac{5}{6}\right) - \frac{1}{6}\log_2\left(\frac{1}{6}\right) \approx 0.65002 \text{ bits}$$Sub-entropy $H(X \mid Y=1)$:$$H(X \mid Y=1) = -\frac{1}{2}\log_2\left(\frac{1}{2}\right) - \frac{1}{2}\log_2\left(\frac{1}{2}\right) = 1.00000 \text{ bit}$$$$H(X \mid Y) = (0.60 \cdot 0.65002) + (0.40 \cdot 1.00000) = 0.39001 + 0.40000 = 0.79001 \text{ bits}$$c) Mutual Information $I(X; Y)$The direct metric reduction in system uncertainty due to shared information:$$I(X; Y) = H(X) - H(X \mid Y) = 0.88129 - 0.79001 = 0.09128 \text{ bits}$$d) Relative Uncertainty Reduction PercentageTo find the efficiency of tracking feature $Y$ over the target field $X$:$$\text{Uncertainty Reduction} = \frac{I(X; Y)}{H(X)} \times 100\% = \frac{0.09128}{0.88129} \times 100\% \approx 10.36\%$$

# Q2) Conditional Density of the Multivariate Gaussian

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

1. Show that the conditional distribution is given by $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right),$$
where
* $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$

Structural Typo Alert in the Prompt ImageBefore starting the formal proof, note a standard typo in the problem sheet's target equation:The normalization factor in the denominator of the provided image is written as $\sqrt{(2\pi)^n|\mathbf{S}_{XX}|}$. In standard literature and by the properties of block matrix operations, this factor must feature the determinant of the conditional covariance matrix $|\mathbf{\Sigma}_{X \mid Y}|$ (which is the Schur complement of $\mathbf{\Sigma}_{YY}$ in $\mathbf{\Sigma}$), rather than a matrix labeled $\mathbf{S}_{XX}$. The formal proof below derives the correct canonical form.
The Identity of the Conditional DensityBy definition of conditional density for absolutely continuous multivariate random variables, we have:$$f_{\mathbf{X} \mid \mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})}$$To find this analytically, we use a linear transformation trick (block diagonalization) to decouple the dependence between $\mathbf{X}$ and $\mathbf{Y}$.

Step 1: Decoupling via Linear TransformationWe define a new random vector $\mathbf{W} \in \mathbb{R}^n$ via a linear transformation that removes the cross-covariance between $\mathbf{X}$ and $\mathbf{Y}$. Let:$$\mathbf{W} = \mathbf{X} - \mathbf{A}\mathbf{Y}$$Where $\mathbf{A} \in \mathbb{R}^{n \times m}$ is a matrix of deterministic coefficients chosen such that $\text{Cov}(\mathbf{W}, \mathbf{Y}) = \mathbf{0}$. We compute this cross-covariance using the linearity of the covariance operator:$$\text{Cov}(\mathbf{W}, \mathbf{Y}) = \text{Cov}(\mathbf{X} - \mathbf{A}\mathbf{Y}, \mathbf{Y}) = \text{Cov}(\mathbf{X}, \mathbf{Y}) - \mathbf{A}\text{Var}(\mathbf{Y}) = \mathbf{\Sigma}_{XY} - \mathbf{A}\mathbf{\Sigma}_{YY}$$Setting this equal to the zero matrix $\mathbf{0}$ yields the optimal projection matrix (the matrix coefficients for linear MMSE estimation):$$\mathbf{\Sigma}_{XY} - \mathbf{A}\mathbf{\Sigma}_{YY} = \mathbf{0} \implies \mathbf{A} = \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}$$Thus, our decoupled transformation variable is:$$\mathbf{W} = \mathbf{X} - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{Y}$$
Step 2: Distributional Parameters of the Transformed SpaceBecause $\mathbf{W}$ and $\mathbf{Y}$ are linear combinations of components of the joint multivariate normal vector $\mathbf{Z}$, the concatenated vector $\begin{bmatrix} \mathbf{W} \\ \mathbf{Y} \end{bmatrix}$ is also jointly multivariate normal.Let us calculate its mean vectors and joint covariance structure:
1. Expectation Vectors$$\mu_W = \mathbb{E}[\mathbf{W}] = \mathbb{E}[\mathbf{X}] - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbb{E}[\mathbf{Y}] = \mu_X - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mu_Y$$$$\mu_Y = \mathbb{E}[\mathbf{Y}]$$
2. Covariance Matrix Elements$\text{Var}(\mathbf{Y}) = \mathbf{\Sigma}_{YY}$$\text{Cov}(\mathbf{W}, \mathbf{Y}) = \mathbf{\Sigma}_{WY} = \mathbf{0}$ (by construction)$\text{Var}(\mathbf{W}) = \mathbf{\Sigma}_{WW}$:$$\mathbf{\Sigma}_{WW} = \text{Cov}(\mathbf{X} - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{Y}, \, \mathbf{X} - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{Y})$$$$\mathbf{\Sigma}_{WW} = \text{Cov}(\mathbf{X},\mathbf{X}) - \text{Cov}(\mathbf{X},\mathbf{Y})\mathbf{\Sigma}_{YY}^{-1}\mathbf{\Sigma}_{YX} - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\text{Cov}(\mathbf{Y},\mathbf{X}) + \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\text{Var}(\mathbf{Y})\mathbf{\Sigma}_{YY}^{-1}\mathbf{\Sigma}_{YX}$$$$\mathbf{\Sigma}_{WW} = \mathbf{\Sigma}_{XX} - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{\Sigma}_{YX} - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{\Sigma}_{YX} + \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{\Sigma}_{YY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{\Sigma}_{YX}$$$$\mathbf{\Sigma}_{WW} = \mathbf{\Sigma}_{XX} - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{\Sigma}_{YX}$$This matrix expression is the Schur complement of $\mathbf{\Sigma}_{YY}$ in $\mathbf{\Sigma}$, which we denote as the conditional covariance matrix:$$\mathbf{\Sigma}_{X \mid Y} = \mathbf{\Sigma}_{XX} - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{\Sigma}_{YX}$$

Step 3: Independence and Conditional ArgumentBecause $\begin{bmatrix} \mathbf{W} \\ \mathbf{Y} \end{bmatrix}$ is jointly Gaussian and their cross-covariance $\mathbf{\Sigma}_{WY} = \mathbf{0}$, uncorrelation implies statistical independence for multivariate normal distributions:$$\mathbf{W} \perp \mathbf{Y}$$Because of this complete independence, conditioning $\mathbf{W}$ on $\mathbf{Y} = \mathbf{y}$ does not alter its probability distribution profile:$$f_{\mathbf{W} \mid \mathbf{Y}}(\mathbf{w} \mid \mathbf{y}) = f_{\mathbf{W}}(\mathbf{w})$$Now, return to the original vector definition $\mathbf{X} = \mathbf{W} + \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{Y}$. If we condition on the event that $\mathbf{Y} = \mathbf{y}$, the term $\mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{Y}$ becomes a constant vector. This means the conditional distribution of $\mathbf{X}$ given $\mathbf{Y}=\mathbf{y}$ is simply a shifted version of the distribution of $\mathbf{W}$:$$\mathbf{X} \mid (\mathbf{Y} = \mathbf{y}) \sim \mathcal{N}\left(\mu_W + \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{y}, \,\, \mathbf{\Sigma}_{WW}\right)$$
Step 4: Final Parametric ConstructionLet us evaluate the specific conditional parameters $\mu^*$ and $\mathbf{\Sigma}^*$ for this distribution:Conditional Mean ($\mu^*$):$$\mu^* = \mu_W + \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{y}$$$$\mu^* = \left(\mu_X - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mu_Y\right) + \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{y}$$$$\mu^* = \mu_X + \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}(\mathbf{y} - \mu_Y)$$Conditional Covariance Matrix ($\mathbf{\Sigma}_{X \mid Y}$):$$\mathbf{\Sigma}_{X \mid Y} = \mathbf{\Sigma}_{WW} = \mathbf{\Sigma}_{XX} - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{\Sigma}_{YX}$$Since a linear shifting transformation preserves the Gaussian nature of a distribution, the conditional distribution $f_{\mathbf{X} \mid \mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$ is a multivariate normal distribution of dimension $n$:$$f_{\mathbf{X} \mid \mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n|\mathbf{\Sigma}_{X \mid Y}|}} \exp\left(-\frac{1}{2}(\mathbf{x} - \mu^*)^T \mathbf{\Sigma}_{X \mid Y}^{-1} (\mathbf{x} - \mu^*)\right)$$Where:$\mathbf{\Sigma}_{X \mid Y} = \mathbf{\Sigma}_{XX} - \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}\mathbf{\Sigma}_{YX}$$\mu^* = \mu_X + \mathbf{\Sigma}_{XY}\mathbf{\Sigma}_{YY}^{-1}(\mathbf{y} - \mu_Y)$This matches the structural framework presented in the problem statement.

# Q3) Continuous Fault Monitoring of a Spindle Motor

A factory monitors the health of a high-speed CNC spindle. The critical variable is the **Internal Temperature ($X$)**, but because the motor is sealed, they can only observe the **Total Spindle Power ($Y$)**.

**The Setup:**
1.  **Hidden Temperature ($X$):** The standardized temperature deviation follows a standard normal distribution:
    $$X \sim \mathscr{N}(0, 1)$$
2.  **Mechanical Load ($Q$):** The torque required to cut material creates "noise" in the power reading. It is independent of temperature and follows:
    $$Q \sim \mathscr{N}(0, 4)$$
3.  **The Measurement ($Y$):** The power sensor $Y$ measures the combined effect of thermal load and mechanical torque:
$$Y = X + Q$$

**Answer the followings:**

**1. Joint Distribution**
Identify the distribution of the random varaible $(X, Y)$.
* Show that $Y \sim \mathscr{N}(0, 5)$.
* Find the covariance $\text{Cov}(X, Y)$.

**2. Information Geometry**
Explain why $X$ is **not** $\sigma(Y)$-measurable. In practical terms, if the power meter shows $Y = 5$, why can't the engineer know for certain if the motor is overheating ($X$ is high) or if the material being cut is simply very hard ($Q$ is high)?

**3. Conditional Expectation (The Best Estimate)**
Compute the conditional expectation $E[X \mid Y]$.
* Using the property for joint normals: $E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$.
* Express this as a "signal-to-total-variance" ratio.

**4. Verification of the Residual**
Let $\widehat{X} = E[X \mid Y]$. Define the estimation error (residual) as $Z = X - \widehat{X}$.
* Prove that $E[Z] = 0$.
* Verify that the error $Z$ is uncorrelated with the observation $Y$ (the Orthogonality Principle).

**5. Information Theory: Differential Entropy**
* Compute the differential entropy $h(X)$.
* Compute the conditional differential entropy $h(X \mid Y)$.
* *Note: The variance of the conditional distribution $(X \mid Y=y)$ is given by $\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)}$.*

**6. Mutual Information**
Calculate $I(X; Y) = h(X) - h(X \mid Y)$.
* What percentage of the "uncertainty" about the temperature is removed by observing the power meter?
* How would this change if the mechanical noise $Q$ had a variance of $100$ instead of $4$?

**7. Diagnostic Interpretation**
The factory manager says: *"Since $Y = X + Q$, our best estimate of temperature should just be $Y$ itself."*
Using your result from Question 3, explain why $Y$ is a **biased** estimator of $X$ and why $E[X \mid Y]$ "shrinks" the measurement toward the mean (zero).



1. Joint Distribution of $(X, Y)$Since $X$ and $Q$ are independent, normally distributed random variables, any linear combination of them is also normally distributed. Because $Y = X + Q$, the vector $\begin{bmatrix} X \\ Y \end{bmatrix}$ is a joint multivariate Gaussian distribution.Proof that $Y \sim \mathcal{N}(0, 5)$Using the linearity of expectation and properties of variance for independent random variables:Mean:$$\mathbb{E}[Y] = \mathbb{E}[X + Q] = \mathbb{E}[X] + \mathbb{E}[Q] = 0 + 0 = 0$$Variance: Since $X \perp Q$, their covariance is zero ($\text{Cov}(X, Q) = 0$):$$\text{Var}(Y) = \text{Var}(X + Q) = \text{Var}(X) + \text{Var}(Q) = 1 + 4 = 5$$Thus, $Y \sim \mathcal{N}(0, 5)$.Covariance Calculation $\text{Cov}(X, Y)$Using the bilinear property of the covariance operator:$$\text{Cov}(X, Y) = \text{Cov}(X, X + Q) = \text{Cov}(X, X) + \text{Cov}(X, Q)$$$$\text{Cov}(X, Y) = \text{Var}(X) + 0 = 1$$Complete Joint Distribution Matrix$$\begin{bmatrix} X \\ Y \end{bmatrix} \sim \mathcal{N}\left( \begin{bmatrix} 0 \\ 0 \end{bmatrix}, \, \begin{bmatrix} 1 & 1 \\ 1 & 5 \end{bmatrix} \right)$$

2. Information Geometry and MeasurabilityFor $X$ to be $\sigma(Y)$-measurable, there must exist a Borel-measurable function $g$ such that $X = g(Y)$ almost surely. This would mean that every state in the underlying sample space mapped to a specific $Y=y$ would yield one—and only one—unambiguous value for $X$.In practical engineering terms, $X$ is not $\sigma(Y)$-measurable because the mapping from $(X, Q) \to Y$ is a many-to-one projection. If the power meter displays $Y = 5$, the underlying state space consists of an infinite line of possibilities satisfying the algebraic constraint $X + Q = 5$. For instance:Scenario A: $X = 4.5$ (severe thermal overheating) and $Q = 0.5$ (idle cutting load).Scenario B: $X = -0.5$ (cool/optimal motor) and $Q = 5.5$ (cutting a highly dense, hard material).Because multiple distinct physical states produce the exact same observable metric $Y = 5$, the engineer cannot invert the observation to isolate $X$ with absolute certainty. The information field $\sigma(Y)$ is a sub-$\sigma$-algebra that blurs these distinctions.
3. Conditional Expectation (The Best Estimate)For a joint multivariate normal distribution, the conditional expectation matches the minimum mean square error (MMSE) linear predictor:$$\mathbb{E}[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$$Substituting our derived values ($\mu_X = 0$, $\mu_Y = 0$, $\text{Cov}(X,Y) = 1$, and $\text{Var}(Y) = 5$):$$\mathbb{E}[X \mid Y = y] = 0 + \frac{1}{5}(y - 0) = \frac{1}{5}y$$Signal-to-Total-Variance RepresentationWe can express this coefficient by substituting the fundamental parameters of the system components ($\text{Cov}(X,Y) = \sigma_X^2$ and $\text{Var}(Y) = \sigma_X^2 + \sigma_Q^2$):$$\mathbb{E}[X \mid Y] = \left( \frac{\sigma_X^2}{\sigma_X^2 + \sigma_Q^2} \right) Y$$This shows that the optimal estimator scales the observation by a signal-to-total-variance ratio (or SNR-like attenuation factor). Here, it yields:$$\mathbb{E}[X \mid Y] = \frac{1}{1 + 4} Y = \frac{1}{5}Y$$
4. Verification of the Residual (The Orthogonality Principle)Let the optimal estimate be $\widehat{X} = \mathbb{E}[X \mid Y] = \frac{1}{5}Y$. The residual error random variable is defined as:$$Z = X - \widehat{X} = X - \frac{1}{5}Y$$Proof that $\mathbb{E}[Z] = 0$Using the linearity of expectation:$$\mathbb{E}[Z] = \mathbb{E}\left[X - \frac{1}{5}Y\right] = \mathbb{E}[X] - \frac{1}{5}\mathbb{E}[Y] = 0 - \frac{1}{5}(0) = 0$$Verification of the Orthogonality Principle ($\text{Cov}(Z, Y) = 0$)To verify that the residual error is completely uncorrelated with our observation space:$$\text{Cov}(Z, Y) = \text{Cov}\left(X - \frac{1}{5}Y, \, Y\right) = \text{Cov}(X, Y) - \frac{1}{5}\text{Var}(Y)$$Substituting our values ($\text{Cov}(X,Y) = 1$ and $\text{Var}(Y) = 5$):$$\text{Cov}(Z, Y) = 1 - \frac{1}{5}(5) = 1 - 1 = 0$$Because $\text{Cov}(Z,Y) = 0$, the error residual carries zero linear information regarding the observation $Y$, confirming the orthogonality of the projection.

Information Theory: Differential EntropyThe differential entropy of a continuous univariate Gaussian distribution $W \sim \mathcal{N}(\mu, \sigma^2)$ is given by:$$h(W) = \frac{1}{2} \ln(2\pi e \sigma^2) \text{ nats}$$1. Differential Entropy $h(X)$Since $\sigma_X^2 = 1$:$$h(X) = \frac{1}{2} \ln(2\pi e (1)) = \frac{1}{2} \ln(2\pi e) \approx 1.4189 \text{ nats}$$2. Conditional Differential Entropy $h(X \mid Y)$The conditional distribution of $X$ given $Y=y$ is also Gaussian. Its conditional variance $\sigma_{X \mid Y}^2$ is computed using the Schur complement formula provided in the note:$$\sigma_{X \mid Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)} = 1 - \frac{(1)^2}{5} = 1 - 0.20 = 0.80$$Because the conditional variance is a deterministic constant that does not depend on the specific value of $y$, the conditional entropy simplifies to:$$h(X \mid Y) = \frac{1}{2} \ln(2\pi e \sigma_{X \mid Y}^2) = \frac{1}{2} \ln(2\pi e \cdot 0.80) \approx 1.3072 \text{ nats}$$
6. Mutual Information AnalysisThe information shared between the hidden temperature and the observable power is:$$I(X; Y) = h(X) - h(X \mid Y) = \frac{1}{2} \ln(2\pi e) - \frac{1}{2} \ln(2\pi e \cdot 0.80)$$$$I(X; Y) = \frac{1}{2} \ln\left( \frac{2\pi e}{2\pi e \cdot 0.80} \right) = \frac{1}{2} \ln\left(\frac{1}{0.80}\right) = \frac{1}{2} \ln(1.25) \approx 0.1116 \text{ nats}$$Percentage of Uncertainty RemovedUsing entropy as the baseline metric for uncertainty, the percentage of information gained by reading the power meter is:$$\text{Percentage} = \frac{I(X;Y)}{h(X)} \times 100\% = \frac{\frac{1}{2}\ln(1.25)}{\frac{1}{2}\ln(2\pi e)} \times 100\% = \frac{\ln(1.25)}{\ln(2\pi e)} \times 100\% \approx \frac{0.22314}{2.83788} \times 100\% \approx 7.86\%$$Case Analysis: Mechanical Noise Variance $\sigma_Q^2 = 100$If the mechanical noise variance scales up to $\sigma_Q^2 = 100$, the system parameters change as follows:$\text{Var}(Y) = \sigma_X^2 + \sigma_Q^2 = 1 + 100 = 101$$\text{Cov}(X,Y) = \sigma_X^2 = 1$The new conditional variance is:$$\sigma_{X \mid Y, \, \text{new}}^2 = 1 - \frac{1^2}{101} = \frac{100}{101} \approx 0.9901$$The new mutual information is:$$I(X; Y)_{\text{new}} = \frac{1}{2} \ln\left(\frac{1}{0.9901}\right) = \frac{1}{2} \ln(1.01) \approx 0.00497 \text{ nats}$$The new percentage of uncertainty removed drops to:$$\text{Percentage}_{\text{new}} = \frac{\ln(1.01)}{\ln(2\pi e)} \times 100\% \approx \frac{0.00995}{2.83788} \times 100\% \approx 0.35\%$$As noise increases, the information channeled through $Y$ drops near zero, making the observation almost entirely uninformative.

Diagnostic InterpretationThe factory manager's intuition ($\widehat{X} = Y$) fails because it ignores the presence of noise, making it a biased estimator when evaluated conditionally.Proof of Conditional BiasAn estimator $\widehat{X}$ is conditionally unbiased if $\mathbb{E}[\widehat{X} \mid X] = X$. If we evaluate the manager's proposal ($\widehat{X} = Y = X + Q$):$$\mathbb{E}[Y \mid X] = \mathbb{E}[X + Q \mid X] = \mathbb{E}[X \mid X] + \mathbb{E}[Q \mid X]$$Since $Q \perp X$, $\mathbb{E}[Q \mid X] = \mathbb{E}[Q] = 0$:$$\mathbb{E}[Y \mid X] = X + 0 = X$$While $Y$ is an unbiased estimator of $X$ on average across the entire joint distribution, the problem is that we do not observe $X$; we observe $Y$. If we evaluate the expectation of $X$ given our actual observation $Y$, the manager's estimate overshoots. The true conditional profile is:$$\mathbb{E}[X \mid Y] = \frac{1}{5}Y$$The Shrinkage PhenomenonThe expression $\mathbb{E}[X \mid Y] = 0.20Y$ demonstrates Bayesian shrinkage. Because the power reading is corrupted by mechanical noise ($\sigma_Q^2 = 4$), an extreme reading (such as $Y = 5$) is statistically more likely to be caused by a large spike in mechanical torque ($Q$) rather than a massive temperature spike ($X$), simply because $Q$ has four times the variance of $X$.The optimal estimator balances our prior information ($\mu_X = 0$) with the noisy measurement ($Y$). To minimize the mean square error, the estimator pulls ("shrinks") the raw measurement toward the prior mean of zero, scaling it down by the factor $\frac{\sigma_X^2}{\sigma_X^2 + \sigma_Q^2} = 0.20$. Using $Y$ directly as the temperature estimate ignores this variance structure and consistently overestimates anomalies.
