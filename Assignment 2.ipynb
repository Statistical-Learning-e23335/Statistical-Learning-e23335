#Q3) Transformer Failiures

In a certain electricity distribution district, the number of emergency transformer failures reported in a single day is modeled as a discrete random variable $X$. Assume that $X$ takes values in the set $\{0,1,2,3\}$, with probabilities

$$
P(X=0)=0.50,\qquad P(X=1)=0.30,\qquad P(X=2)=0.15,\qquad P(X=3)=0.05.
$$

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.
2. Define $X$ as a random variable on this space.
3. Determine the probability mass function of $X$.
4. Compute the expectation $E[X]$.
5. Compute the variance $\mathrm{Var}(X)$.
6. Find $P(X\ge 1)$ and $P(X\ge 2)$.
7. Briefly explain the practical meaning of the expectation and variance in this setting.



1. Construct a Probability Space $(\Omega, \mathcal{F}, P)$A probability space consists of a sample space, a $\sigma$-algebra, and a probability measure.Sample Space ($\Omega$): The set of all possible outcomes. In this case, the number of failures reported.$$\Omega = \{0, 1, 2, 3\}$$$\sigma$-algebra ($\mathcal{F}$): Since the sample space is finite, we typically use the Power Set of $\Omega$, which contains all possible subsets of outcomes.$$\mathcal{F} = \mathcal{P}(\Omega)$$Probability Measure ($P$): A function $P: \mathcal{F} \to [0, 1]$ defined by the given probabilities:$$P(\{0\}) = 0.50, \quad P(\{1\}) = 0.30, \quad P(\{2\}) = 0.15, \quad P(\{3\}) = 0.05$$(Note: These sum to 1.0, satisfying the axiom of total probability.)
2. Define $X$ as a Random VariableA random variable $X$ is a measurable function $X: \Omega \to \mathbb{R}$. For this discrete model, we define the identity mapping:$$X(\omega) = \omega \quad \text{for each } \omega \in \Omega$$This maps the outcomes in our sample space directly to the real numbers representing the count of failures.
3. Probability Mass Function (PMF)The PMF, $p_X(x) = P(X = x)$, is defined for the support $x \in \{0, 1, 2, 3\}$:xP(X=x)00.5010.3020.1530.05
4. Compute the Expectation $E[X]$The expectation (mean) is calculated as $\sum x \cdot P(X=x)$:$$E[X] = (0 \times 0.50) + (1 \times 0.30) + (2 \times 0.15) + (3 \times 0.05)$$$$E[X] = 0 + 0.30 + 0.30 + 0.15 = \mathbf{0.75}$$
5. Compute the Variance $Var(X)$First, we find $E[X^2]$:$$E[X^2] = (0^2 \times 0.50) + (1^2 \times 0.30) + (2^2 \times 0.15) + (3^2 \times 0.05)$$$$E[X^2] = 0 + 0.30 + 0.60 + 0.45 = 1.35$$Now, using the formula $Var(X) = E[X^2] - (E[X])^2$:$$Var(X) = 1.35 - (0.75)^2$$$$Var(X) = 1.35 - 0.5625 = \mathbf{0.7875}$$
6. Find $P(X \geq 1)$ and $P(X \geq 2)$$P(X \geq 1)$: Using the complement rule:$$P(X \geq 1) = 1 - P(X = 0) = 1 - 0.50 = \mathbf{0.50}$$$P(X \geq 2)$: Summing the individual probabilities:$$P(X \geq 2) = P(X = 2) + P(X = 3) = 0.15 + 0.05 = \mathbf{0.20}$$
7. Practical MeaningExpectation ($E[X] = 0.75$): This represents the long-run average number of failures per day. If the district monitors failures over many years, they should expect an average of 0.75 failures daily. It serves as a baseline for resource planning (e.g., staffing repair crews).Variance ($Var(X) = 0.7875$): This measures the uncertainty or spread of the failures. A higher variance would indicate that the number of failures fluctuates wildly from day to day, making scheduling more difficult. Here, the variance is relatively low, suggesting the daily failure rate is somewhat consistent and stays close to the mean.

# Q4) Probability of Defective Items

In a manufacturing process, the proportion of defective items in a large batch produced on a given day is modeled by a continuous random variable $X$. Since $X$ represents a proportion, it takes values in the interval $[0,1]$. Suppose that $X$ has a Beta distribution with parameters $\alpha=2$ and $\beta=18$, so that its density is

$$
f(x)=
\begin{cases}
\dfrac{1}{B(2,18)}x^{1}(1-x)^{17}, & 0<x<1,\\\
0, & \text{otherwise},
\end{cases}
$$

where $B(2,18)$ is the Beta function.

Answer the following:

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.

2. Define $X$ as a continuous random variable on this space.

3. Write down the probability density function of $X$.

4. Verify that $f$ is a valid probability density function.

5. Compute the expectation $E[X]$.

6. Compute the variance $\mathrm{Var}(X)$.

7. Find the probability that the defective proportion is at most $0.10$, that is, compute

$$
P(X\le 0.10).
$$

You may leave your answer in Beta integral form.

8. Find the probability that the defective proportion lies between $0.05$ and $0.15$, that is, compute

$$
P(0.05\le X\le 0.15).
$$

You may leave your answer in integral form.

9. Briefly explain why a Beta distribution is a reasonable model for this situation.

10. Interpret the expectation and variance in the context of production quality.


1. Construct a Probability Space $(\Omega, \mathcal{F}, P)$For a continuous random variable representing a proportion:Sample Space ($\Omega$): The interval of all possible proportions.$$\Omega = [0, 1]$$$\sigma$-algebra ($\mathcal{F}$): The Borel $\sigma$-algebra on $[0, 1]$, denoted $\mathcal{B}([0, 1])$. This contains all intervals and their countable unions/intersections.Probability Measure ($P$): Defined by the integral of the density function $f(x)$ over any set $A \in \mathcal{F}$:$$P(A) = \int_A f(x) dx$$
2. Define $X$ as a Continuous Random Variable$X$ is a measurable function $X: \Omega \to [0, 1]$ such that for every $\omega \in \Omega$, $X(\omega) = \omega$. Because $\Omega$ is an interval and the probability is defined via an integral of a density function, $X$ is continuous.
3. Probability Density Function (PDF)The PDF is given as:$$f(x) = \begin{cases} \frac{1}{B(2, 18)} x^{1}(1-x)^{17}, & 0 < x < 1 \\ 0, & \text{otherwise} \end{cases}$$where the Beta function $B(\alpha, \beta)$ is defined as $\int_0^1 t^{\alpha-1}(1-t)^{\beta-1} dt$.
4. Verify $f$ is a Valid PDFA valid PDF must satisfy two conditions:Non-negativity: $f(x) \geq 0$ for all $x$. Since $x \in (0,1)$, both $x^1$ and $(1-x)^{17}$ are positive. The constant $1/B(2,18)$ is also positive.Normalization: $\int_{-\infty}^{\infty} f(x) dx = 1$.$$\int_0^1 \frac{1}{B(2, 18)} x^{2-1}(1-x)^{18-1} dx = \frac{1}{B(2, 18)} \int_0^1 x^{2-1}(1-x)^{18-1} dx$$By definition of the Beta function, the integral is exactly $B(2, 18)$, so:$$\frac{B(2, 18)}{B(2, 18)} = 1$$
5. Compute the Expectation $E[X]$For a Beta distribution, $E[X] = \frac{\alpha}{\alpha + \beta}$:$$E[X] = \frac{2}{2 + 18} = \frac{2}{20} = \mathbf{0.10}$$
6. Compute the Variance $Var(X)$For a Beta distribution, $Var(X) = \frac{\alpha\beta}{(\alpha+\beta)^2(\alpha+\beta+1)}$:$$Var(X) = \frac{2 \times 18}{(20)^2(21)} = \frac{36}{400 \times 21} = \frac{36}{8400} = \frac{3}{700} \approx \mathbf{0.00428}$$
7. $P(X \leq 0.10)$ in Integral Form$$P(X \leq 0.10) = \int_0^{0.10} \frac{1}{B(2, 18)} x(1-x)^{17} dx$$
8. $P(0.05 \leq X \leq 0.15)$ in Integral Form$$P(0.05 \leq X \leq 0.15) = \int_{0.05}^{0.15} \frac{1}{B(2, 18)} x(1-x)^{17} dx$$
9. Why is the Beta Distribution a reasonable model?The Beta distribution is ideal for modeling proportions or percentages because:Bounded Range: Its support is strictly $[0, 1]$, which matches the physical constraints of a "proportion of defects."Flexibility: By adjusting $\alpha$ and $\beta$, it can take various shapes (skewed, symmetric, or U-shaped). In this case, with $\beta > \alpha$, the distribution is right-skewed, correctly modeling a process where low defect rates are common and high defect rates are rare.
10. Interpretation of ResultsExpectation (0.10): On average, $10\%$ of the items in a batch are expected to be defective. This is the "mean quality level" of the manufacturing process.Variance (0.00428): This represents the consistency of the production quality. A low variance indicates that the proportion of defects doesn't deviate much from the $10\%$ mean from day to day. If the variance were higher, the quality would be considered "unstable," with some days having very high or very low defect rates.

# Q5) Expected Project Completion Time

A construction company is estimating the time required to install a backup generator at a hospital. Let $X$ denote the number of hours needed to complete the installation. Based on expert judgment, the shortest possible time is $4$ hours, the longest possible time is $10$ hours, and the most likely completion time is $6$ hours. Assume that $X$ follows a Triangular distribution with parameters $a=4$, $b=10$, and $c=6$.

Answer the following:

1. Specify the support of $X$.

2. Write down the probability density function $f_X(x)$.

3. Verify that $f_X$ integrates to $1$ over its support.

4. Compute the expectation $E[X]$.

5. Compute the variance $\mathrm{Var}(X)$.

6. Interpret the values of $a$, $b$, and $c$ in the context of this problem.

7. Explain briefly why a Triangular distribution is a reasonable model in this setting.

8. Compute the probability that the installation takes at most $5$ hours, that is, compute

$$
P(X \leq 5).
$$

9. Compute the probability that the installation takes between $5$ and $8$ hours, that is, compute

$$
P(5 \leq X \leq 8).
$$

10. Briefly explain how this model could be used in project planning or sensitivity analysis.


1. Support of $X$The support is the range of values for which the distribution is defined:$$Support(X) = [4, 10]$$
2. Probability Density Function $f_X(x)$The general formula for the Triangular PDF is split into two linear segments:$$f_X(x) = \begin{cases} \frac{2(x-a)}{(b-a)(c-a)} & a \leq x \leq c \\ \frac{2(b-x)}{(b-a)(b-c)} & c < x \leq b \\ 0 & \text{otherwise} \end{cases}$$Plugging in $a=4, b=10, c=6$:For $4 \leq x \leq 6$: $\frac{2(x-4)}{(6)(2)} = \frac{x-4}{6}$For $6 < x \leq 10$: $\frac{2(10-x)}{(6)(4)} = \frac{10-x}{12}$$$f_X(x) = \begin{cases} \frac{x-4}{6}, & 4 \leq x \leq 6 \\ \frac{10-x}{12}, & 6 < x \leq 10 \\ 0, & \text{otherwise} \end{cases}$$
3. Verify $f_X$ Integrates to 1Geometrically, the area of a triangle is $\frac{1}{2} \times \text{base} \times \text{height}$.Base $= b - a = 10 - 4 = 6$Height $= f(c) = \frac{6-4}{6} = \frac{1}{3}$Area $= \frac{1}{2} \times 6 \times \frac{1}{3} = \mathbf{1}$
4. Compute the Expectation $E[X]$The mean of a Triangular distribution is $\frac{a+b+c}{3}$:$$E[X] = \frac{4 + 10 + 6}{3} = \frac{20}{3} \approx \mathbf{6.67 \text{ hours}}$$
5. Compute the Variance $Var(X)$The variance is given by $\frac{a^2 + b^2 + c^2 - ab - ac - bc}{18}$:$$Var(X) = \frac{4^2 + 10^2 + 6^2 - (4 \times 10) - (4 \times 6) - (10 \times 6)}{18}$$$$Var(X) = \frac{16 + 100 + 36 - 40 - 24 - 60}{18} = \frac{28}{18} = \frac{14}{9} \approx \mathbf{1.56}$$
6. Interpretation of $a, b, c$$a = 4$: The optimistic time. The absolute minimum time required if everything goes perfectly.$b = 10$: The pessimistic time. The maximum time required, accounting for potential delays.$c = 6$: The most likely time (mode). The value that would occur most frequently if the installation were repeated many times.
7. Why is the Triangular Distribution a reasonable model?It is a "proxy" distribution used when historical data is scarce. Unlike the Normal distribution, it has finite boundaries (time cannot be negative or infinite), and it allows for asymmetry. In construction, risks usually skew the distribution to the right (the mean $6.67$ is greater than the mode $6.0$), which this model captures.
8. Compute $P(X \leq 5)$Since $5$ is in the first segment ($4 \leq x \leq 6$):$$P(X \leq 5) = \int_4^5 \frac{x-4}{6} dx = \left[ \frac{(x-4)^2}{12} \right]_4^5 = \frac{(1)^2}{12} - 0 = \mathbf{\frac{1}{12} \approx 0.0833}$$
9. Compute $P(5 \leq X \leq 8)$This spans both segments, but it is easier to use $1 - P(X < 5) - P(X > 8)$:$P(X \leq 5) = 1/12$$P(X > 8) = \int_8^{10} \frac{10-x}{12} dx = \left[ -\frac{(10-x)^2}{24} \right]_8^{10} = 0 - \left( -\frac{2^2}{24} \right) = \frac{4}{24} = \frac{1}{6}$$$P(5 \leq X \leq 8) = 1 - \frac{1}{12} - \frac{2}{12} = \frac{9}{12} = \mathbf{0.75}$$
10. Project Planning and Sensitivity AnalysisPlanning: The expectation ($6.67$ hours) is a more realistic figure for scheduling than the mode ($6$ hours) because it accounts for the "long tail" of potential delays.Sensitivity: Managers can perform "What-if" analysis by changing $b$ (e.g., if a part delivery is delayed, $b$ might move to $14$). By observing how the variance and $E[X]$ shift, they can quantify the risk to the hospital's overall timeline.

#Q6) Expected Project Completion Time Using PERT

A project manager is estimating the time required to complete a critical software integration task. Let $X$ denote the number of days required to complete the task. Based on expert judgment, the shortest possible time is $a=8$ days, the longest possible time is $b=20$ days, and the most likely completion time is $c=11$ days. Assume that $X$ follows a PERT distribution with parameters $(a,b,c)$.

The PERT distribution is defined as a Beta distribution scaled to the interval $[a,b]$, with internal shape parameters

$$
\alpha = 1 + 4\left( \frac{c-a}{b-a} \right)
$$

and

$$
\beta = 1 + 4\left( \frac{b-c}{b-a} \right).
$$

Its probability density function is

$$
f(x)=\frac{(x-a)^{\alpha-1}(b-x)^{\beta-1}}{B(\alpha,\beta)(b-a)^{\alpha+\beta-1}}, \qquad a \leq x \leq b,
$$

where

$$
B(\alpha,\beta)=\int_0^1 t^{\alpha-1}(1-t)^{\beta-1},dt.
$$

Answer the following:

1. Compute the values of the shape parameters $\alpha$ and $\beta$.

2. Specify the support of $X$.

3. Write down the probability density function $f_X(x)$ for this model.

4. State the probability space $(\Omega,\mathfrak F,P)$ suitable for this continuous model.

5. Define $X$ as a random variable on this space.

6. Show that the expectation $E[X]$ is given by

$$
E[X]=\frac{a+4c+b}{6}.
$$

7. Compute the variance $\mathrm{Var}(X)$ using the approximation

$$
\mathrm{Var}(X)=\frac{(E[X]-a)(b-E[X])}{7}.
$$

8. Compute the approximate standard deviation $\sigma$.

9. Briefly explain why a PERT distribution may be more appropriate than a Triangular distribution in this setting byt comparing the two results.

10. Interpret the practical meaning of the values $a$, $b$, $c$, $E[X]$, and $\sigma$ in the context of project planning.


1. Compute Shape Parameters $\alpha$ and $\beta$Using the provided formulas:$\alpha = 1 + 4\left(\frac{c-a}{b-a}\right) = 1 + 4\left(\frac{11-8}{20-8}\right) = 1 + 4\left(\frac{3}{12}\right) = 1 + 1 = \mathbf{2}$$\beta = 1 + 4\left(\frac{b-c}{b-a}\right) = 1 + 4\left(\frac{20-11}{20-8}\right) = 1 + 4\left(\frac{9}{12}\right) = 1 + 3 = \mathbf{4}$
2. Support of $X$The support is the range of possible days for the software integration task:$$Support(X) = [8, 20]$$
3. Probability Density Function $f_X(x)$Substituting $\alpha=2$, $\beta=4$, $a=8$, and $b=20$ into the general PERT PDF:$$f_X(x) = \frac{(x-8)^{2-1}(20-x)^{4-1}}{B(2, 4)(20-8)^{2+4-1}} = \frac{(x-8)(20-x)^3}{B(2, 4)(12)^5}$$For $8 \leq x \leq 20$, and $0$ otherwise.
4. State the Probability Space $(\Omega, \mathcal{F}, P)$Sample Space ($\Omega$): The interval $[8, 20]$.$\sigma$-algebra ($\mathcal{F}$): The Borel $\sigma$-algebra $\mathcal{B}([8, 20])$.Probability Measure ($P$): Defined by $P(A) = \int_A f_X(x) dx$ for any $A \in \mathcal{F}$.
5. Define $X$ as a Random Variable$X$ is the identity mapping $X: \Omega \to [8, 20]$ where $X(\omega) = \omega$. It is a continuous random variable because its distribution is defined by a continuous density function.
6. Show the Expectation $E[X] = \frac{a+4c+b}{6}$For a Beta distribution on $[0,1]$, the mean is $\mu_{beta} = \frac{\alpha}{\alpha+\beta}$.The PERT distribution scales this to $[a, b]$ via the transformation $X = a + (b-a)Y$.$$E[X] = a + (b-a)\left(\frac{\alpha}{\alpha+\beta}\right)$$Substitute $\alpha = 1 + 4\frac{c-a}{b-a}$ and $\alpha+\beta = (1 + 4\frac{c-a}{b-a}) + (1 + 4\frac{b-c}{b-a}) = 6$:$$E[X] = a + (b-a)\left(\frac{1 + 4\frac{c-a}{b-a}}{6}\right) = \frac{6a + (b-a) + 4(c-a)}{6} = \frac{6a + b - a + 4c - 4a}{6} = \mathbf{\frac{a + 4c + b}{6}}$$Numerical calculation: $E[X] = \frac{8 + 4(11) + 20}{6} = \frac{72}{6} = \mathbf{12 \text{ days}}$.
7. Compute Variance $Var(X)$Using the provided approximation:$$Var(X) = \frac{(E[X] - a)(b - E[X])}{7} = \frac{(12 - 8)(20 - 12)}{7} = \frac{4 \times 8}{7} = \frac{32}{7} \approx \mathbf{4.57}$$
8. Compute Standard Deviation $\sigma$$$\sigma = \sqrt{Var(X)} = \sqrt{4.5714} \approx \mathbf{2.14 \text{ days}}$$
9. PERT vs. Triangular DistributionThe PERT distribution is often more appropriate because:Smoothness: Unlike the Triangular distribution, which has a "sharp" peak (discontinuity in the derivative at the mode), the PERT distribution is smooth.Weighting: PERT places more emphasis on the "most likely" value ($c$) and less on the extreme outliers ($a$ and $b$), leading to a mean that is less sensitive to extreme pessimistic estimates.
10. Practical Interpretations$a, b, c$: These define the boundaries of risk (8 to 20 days) and the primary focus (11 days).$E[X] = 12$: This is the "weighted average." It tells the manager that despite 11 days being most likely, they should schedule for 12 days to account for the right-hand skew (potential delays).$\sigma = 2.14$: This quantifies the risk. A larger $\sigma$ indicates higher uncertainty in the software integration process. Managers can use $E[X] + \sigma$ to create a "safe" buffer for the project timeline.

#Q7) If the PERT Expectation is to Match Our Intuition

A project manager models the duration $X$ of a task by a PERT distribution with parameters $a$, $b$, and $c$, where $a$ is the minimum time, $b$ is the maximum time, and $c$ is the most likely time.

Suppose the minimum time is $a=5$ days and the most likely time is $c=11$ days. If the expected time computed from the PERT model is also to be the same as the most likely time:

1. Determine the value of the maximum time $b$.

2. What happens to the variance?

1. Determine the Value of the Maximum Time $b$We start with the standard formula for the expectation $E[X]$ of a PERT distribution:$$E[X] = \frac{a + 4c + b}{6}$$The problem states that the expected time must be the same as the most likely time, meaning $E[X] = c$. We can set up the following equation:$$c = \frac{a + 4c + b}{6}$$Now, we solve for $b$:Multiply both sides by 6:$$6c = a + 4c + b$$Subtract $4c$ from both sides:$$2c = a + b$$Isolate $b$:$$b = 2c - a$$Given the values $a = 5$ and $c = 11$:$$b = 2(11) - 5$$$$b = 22 - 5$$$$b = \mathbf{17 \text{ days}}$$Conclusion: For the expectation to match the most likely time of 11 days, the maximum time $b$ must be 17 days.
2. What Happens to the Variance?To understand what happens to the variance, we first look at the shape parameters $\alpha$ and $\beta$.When $E[X] = c$, the distribution is symmetric. We can verify this by calculating $\alpha$ and $\beta$ with our new value $b=17$:$\alpha = 1 + 4\left(\frac{c-a}{b-a}\right) = 1 + 4\left(\frac{11-5}{17-5}\right) = 1 + 4\left(\frac{6}{12}\right) = 1 + 2 = \mathbf{3}$$\beta = 1 + 4\left(\frac{b-c}{b-a}\right) = 1 + 4\left(\frac{17-11}{17-5}\right) = 1 + 4\left(\frac{6}{12}\right) = 1 + 2 = \mathbf{3}$Since $\alpha = \beta$, the distribution is a perfectly symmetric bell-like curve centered at $x = 11$.Impact on Variance Calculation:Using the common PERT variance approximation:$$Var(X) = \frac{(E[X] - a)(b - E[X])}{7}$$Substitute $E[X] = 11$, $a = 5$, and $b = 17$:$$Var(X) = \frac{(11 - 5)(17 - 11)}{7} = \frac{6 \times 6}{7} = \frac{36}{7} \approx \mathbf{5.14}$$Technical Insight: In this symmetric state, the variance reaches its maximum possible value for a fixed range $(b-a)$. If the distribution were skewed (meaning $c$ was closer to $a$ or $b$), the "mass" of the probability would be more bunched up toward one end, typically resulting in a lower variance. By placing the mode exactly in the center, the probability is spread out as widely as possible while maintaining the PERT shape constraints, leading to a relatively higher uncertainty compared to skewed scenarios with the same range.
